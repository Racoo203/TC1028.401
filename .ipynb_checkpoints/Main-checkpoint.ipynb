```python
import random
```

```python
# Función que genera preguntas de figuras retóricas.
def generarPreguntaFiguras():
    # Abrir archivo de ejercicios.
    figuras = open("figuras_Ejercicios.csv","r",encoding="utf-8")
    # Matriz de ejercicios.
    preguntas = []
    
    # Guardar info del archivo en la matriz.
    for linea in figuras:
        # Lista de listas, separados por versos/lineas.
        temp = linea.split("/")
        # Agregar lista a matriz.
        preguntas.append(temp)
    
    # Cerrar archivo de ejercicios.
    figuras.close()

    # Selección de ejercicio; Se escoge un problema aleatorio.
    num = random.randint(0,len(preguntas)-1)
    # Se selecciona el ejercicio con el numero aleatorio.
    ejercicio = preguntas[num]


    # Abrir archivo de respuestas.
    figuras = open("figuras_Respuestas.csv", encoding="utf-8")
    # Matriz de respuestas.
    respuestas = []
    
    # Agrega todas las respuestas a la matriz.
    for linea in figuras:
        # Formato, asegura que solo se tiene el texto y no \n
        txt = linea.replace("\n","")
        # Lista de listas, separados por versos/lineas.
        temp = txt.split("/")
        #
        respuestas.append(temp)

    figuras.close()

    # GUI
    print("Lee el siguiente texto:\n")
    for i in range(len(ejercicio)):
        print(ejercicio[i])

    correcto = False
    
    
    respuesta = input("Indica una figura retórica presente en este texto.\n").lower()

    for i in range(len(respuestas[num])):
        if respuesta == respuestas[num][i].lower():
            correcto = True
    
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~")
 
    if correcto:
        print("CORRECTO! Las posibles respuestas correctas son:")
    else:
        print("Incorrecto, algunas respuestas correctas son:")

    for i in range(len(respuestas[num])):
        print(respuestas[num][i])
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```

```python
def generarPreguntaTexto():
    # Ejercicios de figuras retóricas
    Texto = open("Tipos de Texto P4.csv","r",encoding="utf-8") #Abrirb el docunto con las preguntas
    preguntas = []

    for linea in Texto: #Sacar las preguntas del documento 
        temp = linea.split("/") #Divididrlas si es necesario
        preguntas.append(temp)#Agregar a la varable preguntas

    Texto.close()

    # Selección de ejercicio
    num = random.randint(0,len(preguntas)-1)
    ejercicio = preguntas[num] #De todas las preguntas escoge una aleatoria  


    # Respuestas de los ejercicios
    Texto = open("Texto R.csv", encoding="utf-8") #Para abriri el documento con las respuestas
    respuestas = [] 

    for linea in Texto: #Igual sacar todas las respuestas del documento
        txt = linea.replace("\n","") #Verificar las respuestas y remplazar si es necesario
        temp = txt.split("/")
        respuestas.append(temp) #Agregar las respuestas en una variable

    Texto.close()

    # GUI
    for i in range(len(ejercicio)):# Ya es para enseñar la pregunta
        print(ejercicio[i])

    correcto = False
    
    
    respuesta = input("Indica la respuesta correspondiente:\n").lower() #Input para poner tu respuesta
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~")

    for i in range(len(respuestas[num])): 
        if respuesta == respuestas[num][i].lower(): #Verifica si la respuesta es la misma que la de la pregunta 
            correcto = True #Nos dice que pasa si es correcto

    if correcto:
        print("CORRECTO!") #Imprime Esto cuando es correcto
    else:
        print("Incorrecto, La respuesta correcta es:") #Imprime esto si esta incorrecta

    for i in range(len(respuestas[num])):
        print(respuestas[num][i]) #Aquí ya imprime la respuesta
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```

```python
def generarPreguntaOrtografia():
    
    # Abrimos archivo de preguntas
    figuras = open("Preguntas_Acentos.csv","r",encoding="utf-8")
    #generamos lista de preguntas
    preguntas = []

    #Agregamos las preguntas
    for linea in figuras:

        preguntas.append(linea)

    figuras.close()

    # generamos un número
    num = random.randint(0,len(preguntas)-1)
    
    #Seleccionamos ejercicio con el numero
    ejercicio = preguntas[num]


    # Respuestas de los ejercicios
    
    #Abrimos respuestas
    figuras = open("Respuestas_acentos.csv", encoding="utf-8")
    
    #Generamos lista de respuestas
    respuestas = []
    
    #Limpiamos Archivo
    for linea in figuras:
        txt = linea.replace("\n","")
        respuestas.append(txt)


    figuras.close()

    # GUI
    
    #imprimimos el ejercicio
    
    print("Lee el siguiente texto:\n")
    print(ejercicio)
    
    #generamos variable

    correcto = False
    
    #Preguntamos respuesta
    
    respuesta = input("Indica el acento faltante en este texto.\n")
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~")

    #Comparamos respuesta

    if respuesta == respuestas[num].lower():
        correcto = True

    if correcto:
        print("CORRECTO!")
    else:
        print("Incorrecto, la respuesta correcta es:")

        #Imprimimos respuesta

        print(respuestas[num])
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```

```python
terminar = False

counter = 0;

while(not terminar):
    print("Selecciona que tipo de ejercicio quieres practicar:")
    print("(f) Figuras Retóricas")
    print("(t) Tipos de Textos")
    print("(o) Ortografía")
    print("(e) Cerrar Sesión")
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~")

    opcion = input().lower()

    match opcion:
        case "f":
            generarPreguntaFiguras()
            counter = counter + 1
        case "t":
            generarPreguntaTexto()
            counter = counter + 1
        case "o":
            generarPreguntaOrtografia()
            counter = counter + 1
        case "e":
            terminar = True
            print("Ejercicios hechos:",counter)
        case _:
            print("Selecciona una opcion valida: f, t, o, e.")
            print("~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```
